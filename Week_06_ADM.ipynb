# *Q1: Load dataset and display first three rows*



```python
import pandas as pd
data = {
    "Order_ID":[2005,2006,2003,2008],
    "Product":["Refridgerator","Airpods","Keyboard","Mouse"],
    "Price($)":[2000, None, 650, 1400],
    "Quantity":[3, 4, None, 2],
    "Discount(%)":[20, 15, 9, None]
}
df = pd.DataFrame(data)
print(df.head(3))
```

# Q2: Identify missing values and count them

```python
print(df.isnull().sum()) # Counts missing values
```

## Q3. Fill missing Price with mean and missing Quantity with 1


```python
df['Price($)'].fillna(df['Price($)'].mean(),inplace=True)
df['Quantity'].fillna(1,inplace=True)
```

# Q4: Calculate total revenue

```python
df['Revenue'] = (df['Price($)'] * df['Quantity']) - df['Discount(%)'].fillna(0) # Replaced 'Price ($)' with 'Price($)' and 'Discount (%)' with 'Discount(%)'
print(df[['Order_ID','Revenue']])
```

# Q5: Why is data cleaning important in e-commerce?

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(df["Price($)"], bins=5, kde=True, color="blue")
plt.title("Distribution of Product Prices")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Product", y="Quantity", data=df, palette="magma")
plt.title("Quantity Sold Per Product")
plt.xticks(rotation=45)
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Price($)", y="Discount(%)", data=df, color="Purple", s=100)
plt.title("Discount Percentage vs. Product Price")
plt.xlabel("Price ($)")
plt.ylabel("Discount (%)")
plt.show()
```
